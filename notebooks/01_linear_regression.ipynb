```python
#Import Libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the Dataset
df = pd.read_csv('../datasets/Salary_dataset.csv')
print(f"Number of training examples: {len(df)}")
x = df['YearsExperience'].values
y = df['Salary'].values
print(f"Example Datapoint: x₀ = {x[0]}, y₀ = {y[0]}")
```

```python
# Plot the data
plt.scatter(x,y)
plt.title("Salary dataset")
plt.ylabel("Salary")
plt.xlabel("Years of Experience")
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      f_wb (ndarray (m,)): model prediction
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
        
    return f_wb
```
